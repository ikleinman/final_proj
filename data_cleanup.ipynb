```python
print("Importing dependencies")
import os
import csv
import pandas as pd
import numpy as np
from functools import reduce
```

```python
print("Read csv of identified features")
features = pd.read_csv("Compiled_Features.csv")
features.head()
```

```python
print("Turn features into list")
feature_list = features["Code"].tolist()
```

```python
print("Create list of csvs for column identification")
csv_list = ["original_csv_files/demographic.csv", 
           "original_csv_files/diet.csv",
          "original_csv_files/examination.csv",
          "original_csv_files/labs.csv",
          "original_csv_files/medications.csv",
          "original_csv_files/questionnaire.csv"]
csv_list
```

```python
print("Identify csvs containing identified features, with corrosponding feature names")
print("-------------------------------------------------------------------------------")
for csv in csv_list:
    usecols = []
    csv_df = pd.read_csv(csv, nrows=1)
    csv_col_list = csv_df.columns.tolist()
    for feature in feature_list:
        if feature in csv_col_list:
            usecols.append(feature)

    print(csv + " contains " + str(len(usecols)) + " features.")
    print(usecols)
```

```python
print("Read in demographic csv with relevant features")
usecols = []
demographic = pd.read_csv("original_csv_files/demographic.csv", nrows=1)
demographic_list = demographic.columns.tolist()
for feature in feature_list:
    if feature in demographic_list:
        usecols.append(feature)
    demographic = pd.read_csv("original_csv_files/demographic.csv", usecols=usecols)
demographic.head()
```

```python
print("Read in diet csv with relevant features")
usecols = []
diet = pd.read_csv("original_csv_files/diet.csv", nrows=1)
diet_list = diet.columns.tolist()
for feature in feature_list:
    if feature in diet_list:
        usecols.append(feature)
    diet = pd.read_csv("original_csv_files/diet.csv", usecols=usecols)
diet.head()
```

```python
print("Read in examination csv with relevant features")
usecols = []
examination = pd.read_csv("examination.csv", nrows=1)
examination_list = examination.columns.tolist()
for feature in feature_list:
    if feature in examination_list:
        usecols.append(feature)
    examination = pd.read_csv("examination.csv", usecols=usecols)
examination.head()
```

```python
print("Read in questionnaire csv with relevant features")
usecols = []
questionnaire = pd.read_csv("questionnaire.csv", nrows=1)
questionnaire_list = questionnaire.columns.tolist()
for feature in feature_list:
    if feature in questionnaire_list:
        usecols.append(feature)
    questionnaire = pd.read_csv("questionnaire.csv", usecols=usecols)
questionnaire.head()
```

```python
print("Delete SLQ050 nan rows")
questionnaire = questionnaire.dropna(subset=["SLQ050"])
questionnaire.head()
print("There are " + str(len(questionnaire.SEQN)) + " rows remaining")
```

```python
print("Merge dataframes")
df = questionnaire.merge(demographic, how="left", on="SEQN")
df = df.merge(diet, how="left", on="SEQN")
df = df.merge(examination, how="left", on="SEQN")

df.head()
```

```python
print("Move target feature to end of dataframe")
cols_at_end = ["SLQ050"]
df = df[[c for c in df if c not in cols_at_end] + [c for c in cols_at_end if c in df]]
df.head()
```
